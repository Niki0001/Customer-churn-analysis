```python
import pandas as pd

# Dataset load karo
df = pd.read_csv('ecommerce_customer_churn_data.csv')

# Pehle 5 rows dekho
print(df.head())

# Shape
print("Rows & Columns:", df.shape)

# Column names
print("Columns:", df.columns.tolist())

# Basic info
print(df.info())
```

```python
import matplotlib.pyplot as plt

# 1. Churn rate kitni hai overall
churn_counts = df['Is_Churn'].value_counts()
churn_percent = df['Is_Churn'].value_counts(normalize=True) * 100

print("Churn Count:")
print(churn_counts)
print("\nChurn Percentage:")
print(churn_percent.round(2))

# 2. Churn pie chart
plt.figure(figsize=(6,6))
plt.pie(churn_counts, 
        labels=['Not Churned', 'Churned'], 
        autopct='%1.1f%%',
        colors=['#2ecc71', '#e74c3c'])
plt.title('Overall Customer Churn Rate')
plt.savefig('churn_rate.png')
plt.show()

# 3. Contract Type vs Churn
contract_churn = df.groupby('Contract_Type')['Is_Churn'].mean() * 100
print("\nChurn Rate by Contract Type:")
print(contract_churn.round(2))
```

```python
# 1. Average age - churned vs not churned
print("Average Age:")
print(df.groupby('Is_Churn')['Age'].mean().round(2))

# 2. Satisfaction Score vs Churn
print("\nAvg Satisfaction Score:")
print(df.groupby('Is_Churn')['Satisfaction_Score'].mean().round(2))

# 3. Customer Support Calls vs Churn
print("\nAvg Support Calls:")
print(df.groupby('Is_Churn')['Customer_Support_Calls'].mean().round(2))

# 4. Bar charts - teeno ek saath
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Age
df.groupby('Is_Churn')['Age'].mean().plot(
    kind='bar', ax=axes[0], color=['#2ecc71','#e74c3c'])
axes[0].set_title('Avg Age vs Churn')
axes[0].set_xticklabels(['Not Churned','Churned'], rotation=0)

# Satisfaction
df.groupby('Is_Churn')['Satisfaction_Score'].mean().plot(
    kind='bar', ax=axes[1], color=['#2ecc71','#e74c3c'])
axes[1].set_title('Satisfaction Score vs Churn')
axes[1].set_xticklabels(['Not Churned','Churned'], rotation=0)

# Support Calls
df.groupby('Is_Churn')['Customer_Support_Calls'].mean().plot(
    kind='bar', ax=axes[2], color=['#2ecc71','#e74c3c'])
axes[2].set_title('Support Calls vs Churn')
axes[2].set_xticklabels(['Not Churned','Churned'], rotation=0)

plt.tight_layout()
plt.savefig('churn_analysis.png')
plt.show()
```

```python
# Pehle print karo
print("Avg Monthly Spend:")
print(df.groupby('Is_Churn')['Monthly_Spend'].mean().round(2))

print("\nAvg Last Purchase Days Ago:")
print(df.groupby('Is_Churn')['Last_Purchase_Days_Ago'].mean().round(2))

print("\nAvg Monthly Logins:")
print(df.groupby('Is_Churn')['Monthly_Logins'].mean().round(2))

# Heatmap - sirf numeric columns
import seaborn as sns

numeric_df = df.select_dtypes(include='number')  # sirf numbers lo

plt.figure(figsize=(10, 8))
sns.heatmap(numeric_df.corr(), 
            annot=True, 
            fmt='.2f',
            cmap='RdYlGn',
            center=0)
plt.title('Correlation Heatmap — Churn Analysis')
plt.tight_layout()
plt.savefig('correlation_heatmap.png')
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Customer Churn Analysis — Key Insights', 
             fontsize=16, fontweight='bold')

# 1. Contract Type vs Churn
contract_churn = df.groupby('Contract_Type')['Is_Churn'].mean() * 100
contract_churn.plot(kind='bar', ax=axes[0,0], 
                    color=['#2ecc71','#e74c3c'])
axes[0,0].set_title('Churn Rate by Contract Type')
axes[0,0].set_ylabel('Churn Rate (%)')
axes[0,0].set_xticklabels(['Annual','Monthly'], rotation=0)
for i, v in enumerate(contract_churn):
    axes[0,0].text(i, v+0.2, f'{v:.1f}%', ha='center', fontweight='bold')

# 2. Satisfaction Score vs Churn
sat_churn = df.groupby('Is_Churn')['Satisfaction_Score'].mean()
sat_churn.plot(kind='bar', ax=axes[0,1],
               color=['#2ecc71','#e74c3c'])
axes[0,1].set_title('Avg Satisfaction Score vs Churn')
axes[0,1].set_ylabel('Satisfaction Score')
axes[0,1].set_xticklabels(['Not Churned','Churned'], rotation=0)
for i, v in enumerate(sat_churn):
    axes[0,1].text(i, v+0.05, f'{v:.2f}', ha='center', fontweight='bold')

# 3. Support Calls vs Churn
calls_churn = df.groupby('Is_Churn')['Customer_Support_Calls'].mean()
calls_churn.plot(kind='bar', ax=axes[1,0],
                 color=['#2ecc71','#e74c3c'])
axes[1,0].set_title('Avg Support Calls vs Churn')
axes[1,0].set_ylabel('Support Calls')
axes[1,0].set_xticklabels(['Not Churned','Churned'], rotation=0)
for i, v in enumerate(calls_churn):
    axes[1,0].text(i, v+0.05, f'{v:.2f}', ha='center', fontweight='bold')

# 4. Last Purchase Days vs Churn
purchase_churn = df.groupby('Is_Churn')['Last_Purchase_Days_Ago'].mean()
purchase_churn.plot(kind='bar', ax=axes[1,1],
                    color=['#2ecc71','#e74c3c'])
axes[1,1].set_title('Avg Days Since Last Purchase vs Churn')
axes[1,1].set_ylabel('Days')
axes[1,1].set_xticklabels(['Not Churned','Churned'], rotation=0)
for i, v in enumerate(purchase_churn):
    axes[1,1].text(i, v+0.2, f'{v:.1f}', ha='center', fontweight='bold')

plt.tight_layout()
plt.savefig('churn_dashboard.png', dpi=150, bbox_inches='tight')
plt.show()
print("Dashboard saved!")
```
